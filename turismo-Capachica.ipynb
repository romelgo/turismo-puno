```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np # Necesario para cálculos de posición de barras


# --------------------------------------------------------------------------
# PASO 2: Definir la asociación y filtrar los datos (Sin cambios)
# --------------------------------------------------------------------------
nombre_asociacion_especifica = 'Asociacion Comunal de Turismo Ccotos'

df_filtrado_asociacion = df_long[df_long['ASOCIACION'] == nombre_asociacion_especifica].copy()

if df_filtrado_asociacion.empty:
    print(f"ERROR: No se encontraron datos para la asociación '{nombre_asociacion_especifica}'.")
    raise SystemExit()
else:
    print(f"Datos encontrados para '{nombre_asociacion_especifica}':")
    # Ordenar por Año y Tipo para asegurar consistencia
    df_filtrado_asociacion = df_filtrado_asociacion.sort_values(by=['Año', 'TIPO'])
    print(df_filtrado_asociacion)
    print("\n" + "="*50 + "\n")

# --------------------------------------------------------------------------
# PASO 3: Preparar datos y crear la Gráfica de Barras Agrupadas
# --------------------------------------------------------------------------

# --- 3.1: Pivotar los datos para tener Tipos como columnas ---
# Esto facilita la creación de barras agrupadas
df_pivot = df_filtrado_asociacion.pivot(index='Año', columns='TIPO', values='Valor').fillna(0)
# Asegurarse que las columnas esperadas (NACIONAL, EXTRANJERO) existen, si no, añadir con 0
for tipo in ['NACIONAL', 'EXTRANJERO']:
    if tipo not in df_pivot.columns:
        df_pivot[tipo] = 0
df_pivot = df_pivot[['NACIONAL', 'EXTRANJERO']] # Asegurar orden de columnas

print("--- Datos Pivotados para Gráfico ---")
print(df_pivot)
print("\n" + "="*50 + "\n")

# --- 3.2: Configuración del gráfico ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 7)) # Ajustar tamaño si es necesario

# --- 3.3: Calcular posiciones de las barras ---
años = df_pivot.index.to_numpy() # Años en el eje X
n_tipos = len(df_pivot.columns) # Número de tipos (Nacional, Extranjero) -> 2
bar_width = 0.35 # Ancho de cada barra individual
index = np.arange(len(años)) # Posiciones base para cada grupo de barras (cada año)

# Calcular la posición x para cada tipo de barra dentro de su grupo anual
posiciones = {}
for i, tipo in enumerate(df_pivot.columns):
    # Calcular desplazamiento desde el centro del grupo (index)
    desplazamiento = (i - (n_tipos - 1) / 2) * bar_width
    posiciones[tipo] = index + desplazamiento

# --- 3.4: Dibujar las barras ---
colores = {'NACIONAL': 'tab:blue', 'EXTRANJERO': 'tab:orange'}

for tipo in df_pivot.columns:
    ax.bar(posiciones[tipo], df_pivot[tipo], bar_width,
           label=tipo, color=colores.get(tipo, 'gray'))

# --- 3.5: Títulos y Etiquetas ---
ax.set_title(f'Flujo Anual de Visitantes (Nacionales vs. Extranjeros) - 2018-2023\n{nombre_asociacion_especifica} \n Capachica',
             fontsize=14, pad=15)
ax.set_xlabel('Año', fontsize=12)
ax.set_ylabel('Número de Visitantes', fontsize=12)

# --- 3.6: Ajustes de Ejes ---
ax.set_xticks(index) # Poner las etiquetas de año en el centro de cada grupo de barras
ax.set_xticklabels(años)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Ajustar límite inferior del eje Y a 0 para barras
max_y = df_pivot.max().max() # El valor más alto en todo el pivot
ax.set_ylim(bottom=0, top=max_y * 1.15) # Empezar en 0, dejar espacio arriba

# --- 3.7: Leyenda ---
ax.legend(title='Tipo de Visitante', title_fontsize='11', fontsize='10')

# --- 3.8: Cuadrícula (solo en eje Y para claridad) ---
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.xaxis.grid(False) # Quitar cuadrícula vertical

# --- 3.9 (Opcional): Añadir valores encima de las barras ---
for tipo in df_pivot.columns:
    for i, valor in enumerate(df_pivot[tipo]):
        if valor > 0: # Solo mostrar etiqueta si el valor es mayor que 0
             ax.text(posiciones[tipo][i], valor + max_y*0.01, f'{int(valor):,}', # Texto con formato
                     ha='center', va='bottom', fontsize=8, color=colores.get(tipo, 'gray'))


# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
